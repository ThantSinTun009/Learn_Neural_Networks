# Training Sparse Models: Why and How

In most machine learning tasks, the optimization algorithms we use—like SGD, Adam, or RMSProp—tend to produce dense models. This means that after training, almost all the model parameters (weights) are nonzero. While dense models often achieve high accuracy, they come with two drawbacks:

1. Memory usage: Dense models require more storage since every parameter, no matter how small, takes up space.

2. Computational efficiency: More nonzero weights mean more multiplications during inference, slowing down predictions.

For applications where speed and memory are critical—such as mobile devices, embedded systems, or real-time predictions—it is advantageous to have **a sparse model**, where most of the weights are exactly zero.

> Sparse models are smaller, faster, and easier to deploy.

image.png

A simple way to create a sparse model is to train a normal dense model first and then **manually set very small weights to zero** (get rid of the tiny
weights).

However, this will typically not lead to a very sparse model,
and it may degrade the model’s performance.

### Using L1 regularization (Lasso Regression)
A better option is to apply **strong ℓ1 regularization** during training, as it pushes the optimizer to zero out as many weights as it can. Unlike ℓ2 regularization, which shrinks weights toward zero but rarely makes them exactly zero, ℓ1 regularization adds a penalty proportional to the absolute value of each weight.

> However, in some cases these techniques may remain insufficient.
One last option is to apply *Dual Averaging*, often called Follow **The Regularized Leader (FTRL)**, a technique proposed by Yurii Nesterov.

When used with ℓ1 regularization, this technique often leads to very sparse models.

Keras implements a variant of FTRL called **FTRLProximal21** in the FTRL optimizer.
