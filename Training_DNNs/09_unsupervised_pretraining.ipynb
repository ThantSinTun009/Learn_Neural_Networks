Sometimes you want to solve a difficult machine learning task, but you have very little labeled data.

Labeling data is expensive and slow, but unlabeled data is usually easy to get.

When this happens, **unsupervised pretraining** can help.

If you can gather plenty of unlabeled training data:

image.png

### **The Core Idea**

#### 1. Train the model without labels first (with autoencoders or RBMs)
- Use a lot of unlabeled data
- Let the model learn patterns and features by itself
- **Learn Low-Level Features (Unsupervised)**

#### 2. Train layer by layer
- Start with the first (lowest) layer
- Freeze it once trained (**become feature extractor**)

#### 3. Stack Another Layer
- Train the next layer using the output of the previous one
- Learns higher-level patterns
- Repeat until all layers are trained

```python
Low layers â†’ simple features

Middle layers â†’ abstract concepts

High layers â†’ task-relevant patterns
```

#### 4. Use unsupervised methods
- Common techniques:
  - Autoencoders (most common today)
  - Restricted Boltzmann Machines (RBMs) (used more in the past)

#### 5. Add labels at the end
- After pretraining, **add the Task-Specific Output Layer**
- The network already understands the data structure
- Only needs to map learned features to labels

#### 6. Fine-Tuning (Supervised Learning)
- unfreeze some or all layers
- The entire network is trained end-to-end
- model starts from a good position, not random noise.

> This is a rather long and tedious process, but it often works well; in fact, it is this technique that *Geoffrey Hinton* and his team used in 2006, which led to the **revival of neural networks** and the success of Deep Learning.



Before 2010, unsupervised pretraining was very common.

Later, better initialization and activation functions reduced the need for it

Still useful today when labeled data is limited
