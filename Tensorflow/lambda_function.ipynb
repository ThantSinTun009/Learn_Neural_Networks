# **Explore lambda functions**

**Structure**

```python
lambda param(s) : operation
```

- No return statement

- No Name (Anonymous function)

````python
lambda x: x ** 2
````

````python
lambda num_1, num_2: num_1+num_2
````

````python
def square(n):
  return n**2

def cube(n):
  return n**3

# higher-order function      # simple helper function
def transform_list(num_list, transform_item):
  transfrom_0 = transform_item(num_list[0])
  transfrom_1 = transform_item(num_list[1])
  return [transfrom_0, transfrom_1]

transform_list([2, 3], cube)
````

````python
transform_list([4, 5], square)
````

````python
transform_list([2, 3], lambda x: x + 5)
````

## Python higher-order functions

### map( )

```python
list(map(function, list))

````python
num_list = [1, 2, 3, 4, 5, 8, 9]

list(map(lambda x: x + 7, num_list))
````

### filter( )

Keep only items that meet conditions

````python
list(filter(lambda x: x % 2 == 0, num_list))
````
