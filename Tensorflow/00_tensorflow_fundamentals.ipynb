### **Creating random tensors**

Random tensors are tensors of some arbitrary size which contain random numbers.

```python
random_1 = tf.random.Generator.from_seed(7)
random_1 = random_1.normal(shape=(3, 2))
random_1

random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3, 2))
random_2
```

### Shuffle the order of elements in a tensor

```python
# shuffle tensor (valuable for when you want to shuffle your data so the inherent order doesn't effect learning)

not_shuffled = tf.constant([[10, 7],
                      [3, 4],
                       [2, 5]])
# shuffle this
tf.random.shuffle(not_shuffled)
```

```python
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled, seed=9)
```

```python
ts1 = tf.random.Generator.from_seed(9)
ts1 = tf.random.normal(shape=[2, 3])
ts1
```

```python
tf.random.set_seed(42)
tf.random.shuffle(ts1, seed=7)
```

```python
tf2=tf.random.Generator.from_seed(7)
tf2 = tf.random.normal(shape=[2,2])

print(tf2)

tf.random.set_seed(7)
print(tf.random.shuffle(tf2, seed=9))
```

```python
tensor = tf.random.Generator.from_seed(9)
tensor = tf.random.normal(shape=[3, 4])

print(tensor)

tf.random.set_seed(8)
print(tf.random.shuffle(tensor, seed=2))
```

# Getting information from tensors

When dealing with tensors you probably want to be aware of the following attributes:

Shape  
Rank    
Axis or dimension   
Size

```python
# create rank 4 tensor (4D)
rank_4_tensor = tf.zeros([2, 3, 4, 5])
rank_4_tensor
```

```python
print(rank_4_tensor.shape)
print('\n')

print(rank_4_tensor.ndim)
print('\n')

print(tf.size(rank_4_tensor))
```

```python
2 * 3 * 4 * 5
```

# Various attributes of our tensor

```python
print('Datatype of every elements : ', rank_4_tensor.dtype)
print(f'Number of dimensions (rank) : {rank_4_tensor.ndim}')
print('Shape of tensor : ', rank_4_tensor.shape)
print('Elemeents along the 0 axis: ', rank_4_tensor.shape[0])
```

```python
print('Elements along the last axis : ', rank_4_tensor.shape[-1])
print('Total number of elements in our tensor: ', tf.size(rank_4_tensor))
print('Total number of elements in our tensor : ', tf.size(rank_4_tensor).numpy())
```

### Matrix multiplication

In machine learning, matrix multiplication is one of the most common tensor operations.

There are two rules our tensors (or matrices) need to fulfil if we're going to matrix multiply them:

- The inner dimensions must match
- The resulting matrix has the shape of the outer dimensions

```python
# Matrix multiplication in tensorflow
print(tensor)
tf.matmul(tensor, tensor)
```

```python
tensor, tensor
```

```python
tensor * tensor
```

```python
# Matrix mutliplication with Python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create 3,2 shape matrices
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])

Y = tf.constant([[7,8],
                 [9, 10],
                [11, 12]])

X, Y
```

```python
tf.matmul(X, Y)
```

```python
# change shape of Y
tf.reshape(Y, shape=[2, 3])
```

```python
X.shape, tf.reshape(Y, shape=[2,3])
```

```python
X @ tf.reshape(Y, shape=[2,3])
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
tf.reshape(X, shape=(2, 3)).shape, Y.shape
```

```python
# Try change the shape of X instead of Y
tf.reshape(X, shape=(2, 3)) @ Y
```

**Can use Transpose**

```python
X, tf.transpose(X), tf.reshape(X, shape=[2,3])
```

```python
tf.matmul(tf.transpose(X), Y)
```

ðŸ“– Resource: Info and example of matrix multiplication: https://www.mathsisfun.com/algebra/matrix-multiplying.html

```python
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
tf.matmul(X, tf.transpose(Y))
```

```python

# Check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # "\n" is for newline

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will tranpose (rather than reshape) one of the tensors to get satisify the matrix multiplication rules.

# Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values.

```python
D = tf.constant([-7, -10])
D
```

```python
# get absolute values
tf.abs(D)
```

Let's go through the following forms of aggregation:

- Get the minimum
- Get the maximum
- Get the mean of a tensor
- Get the sum of a tensor

```python
E = tf.constant(np.random.randint(0, 100, size=50))
E
```

```python
tf.size(E), E.shape, E.ndim
```

```python
# find minimum
tf.reduce_min(E)
```

```python
# maximum
tf.reduce_max(E)
```

```python
# mean
tf.reduce_mean(E)
```

```python
# Sum
tf.reduce_sum(E)
```

```python
# standard deviation
tf.reduce_variance(E)
```

```python
# for variance, need to use tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# find standard deviation
tf.math.reduce_std(tf.cast(E, dtype=tf.float32))
```

```python
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```
